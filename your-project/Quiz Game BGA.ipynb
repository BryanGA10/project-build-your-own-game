```python
import random
from string import ascii_lowercase

Num_questions_per_quiz = 5
Questions = { 
    "While fighting his own father in the Cloud City, which hand does Luke Skywalker lose": [
        'Right', 'Left', 'Both', 'Neither'
    ], "What is Chewbacca's weapon of choice": [
        "Bowcastrer", "lightsaber", "Metal stick", "Blaster"
    ], "Where did the Clone Wars start": [
        "Geonosis", "Tatooine", "Naboo", "coruscant"
    ], "What is the name of the sith lord who has a red double lightsaber": [
        'Darth Maul', 'Darth Vader', 'Darth Mulk', 'Darth Garth'
    ], "Who adopted Padm√© Amidala's daughter": [
        'Bail Organa', 'Capit√°n Antilles', 'Owen y Beru Lars', 'Gidean Danu'
    ],  "Who is Rey's grandfather": [
        'Palpatine', 'Anakin Skywalker', 'Obi Wan Kenobi', 'Padme Amidala'
    ],  "Who freed Princess Leia from her chains in Jabba's palace": [
        'R2-D2', 'C3-PO', 'BB-8', 'R4-D4'
    ],  "What is the name of Samuel L. Jackson's character": [
        'Mace Windu', 'Bail Organa', 'Grogu', 'Mas Kanata'
    ],  "What color is Qui-Gon Jinn's lightsaber": [
        'Green', 'Red', 'Blue', 'Purple'
    ], "While the Jedi use a number of crystal colors for their lightsabers, what color is preferred by the Sith": [
        'Red', 'Purple', 'Blue', 'Green'
    ]  
}

Questions_master = {
    "Which Star Wars film has this quote: They fly now": [
        'The Rise of Skywalker', 'Attack of the clones', 'Rogue One', 'Solo'
    ],  "What kind of vehicle did Rey live in": [
        'AT-AT', 'Star Destroyer', 'Mon Calimari', 'AT-ST'
    ],  "Which Jedi says this quote: Fear is the path to the dark side": [
        'Yoda', 'Luke Skywalker', 'Obi-Wan Kenobi', 'Kanan Jarrus'
    ] , "Which Jedi is corrupted on Mortis, is falsely accused, and is snippy": [
        'Ahsoka Tano', 'Anakin Skywalker', 'Obi Wan Kenobi', 'Jar Jar Binks'
    ],   "Where does Yoda live when he first trains Luke Skywalker": [
        'Dagobah', 'Hoth', 'Coruscant', 'Tatoonie'
    ], "While fighting his own father in the Cloud City, which hand does Luke Skywalker lose": [
        'Right', 'Left', 'Both', 'Neither'
    ], "Who defeats Finn in his very first lightsaber duel": [
        'Kylo Ren', 'Darth Maul', 'Darth Vader', 'Kerok Vet'
    ], "What was the job that Finn told Han Solo that he had at Starkiller Base": [
        'Guard', 'Pilot', 'Cook', 'Soldier'
    ],   "What are the creatures that live on Endor that helped the rebels defeat the second Death Star": [
        'Ewoks', 'Wookies', 'Nerf Herders', 'Jawas'
    ], "How many engines does an X-Wing fighter have": [
        '4', '6', '2', '8'
    ]
}

Questions_ghost = {
    "Which Star Wars film has this quote: Always remember, your focus determines your reality": [
        "The Phantom Menace", "The Empire Strikes Back", "The Force Awakens", 
"Rogue One"
    ], "Which stormtrooper wasn't able to complete his mission in Star Wars: The Force Awakens": [
        "FN-2187", "FN-2705", "FN-1205", "FN-1312"
    ], "Which Jedi hates sand, loves Padm√©, and is too old to train": [
        "Anakin Skywalker", "Dooku", "Yoda Master", "Boba Fett"
    ], "In The Force Awakens, which character has Darth Vader‚Äôs damaged mask": [
        'Kylo Ren', 'Finn', 'Rey', 'Luke Skywalker'
    ], "Who killed Jabba": [
        'Princess Leia', 'Han Solo', 'C-3PO', 'Luke Skywalker'
    ], "What's the name of the most sarcastic droid ever created": [
        'K-2SO', 'BB-8', 'R4-D4', 'DAVE'
    ], "According to Master Yoda, how many Sith are always out there": [
        '2', '3', '1', '4'
    ], "Jyn Erso appeared in which Star Wars film": [
        'Rogue One', 'The Empire Strikes Back', 'Return of the Jedi', 'The Phantom Menace'
    ], "Who destroyed the Droid Control Ship when they accidentally activated a Naboo Fighter‚Äôs Autopilot": [
        'Anakin Skywalker', 'Obi Wan Kenobi', 'Jar Jar Binks', 'Biggs Darkligther'
    ], "Who wiped out all of the information about Kamino from the Jedi Temple archives": [
        'Count Dooku', 'Yoda', 'Sifo-Dyas', 'Darth Sidious'
    ]   
}

def run_quiz():
    
        
    print('ü™ê STAR WARS QUIZ ü™ê \n')
    user_name = input("Enter your name: ")
    print(f"\n Welcome {user_name}, Let's see if the force is strong with you!")
    
    
     #Preprocess
    
    play = "a"
    
    while play == "a":
        
        Level = get_level()

        if Level == 'a':
            questions = prepare_questions(
            Questions, num_questions=Num_questions_per_quiz)
        elif Level == 'b':
            questions = prepare_questions(
            Questions_master, num_questions=Num_questions_per_quiz)
        else:
            questions = prepare_questions(
            Questions_ghost, num_questions=Num_questions_per_quiz)


        #Process (main loop)
        num_correct = 0
        for num, (question, alternatives) in enumerate(questions, start=1):
            print(f"\nQuestion {num}: ")
            num_correct +=ask_question(question, alternatives)

        #Postprocess
        print(f"\nYou got {num_correct} correct out of {num} questions")
        
        play = play_again()
        
        if play == 'b':
            print('Thanks fo playing!')

def play_again():
    
    level_options = ['a','b']
    
    print("""\n Do you wanna play again?: 
    
    a) Yes
    b) No""")
        
    while (play := input("\nChoice? "))  not in level_options:
        print(f"Please select one of the options {', '.join(level_options)}")
        
    return play
    
    
def get_level():
   
    levels = ['a', 'b', 'c']
    print("""\n Select the difficulty level: 
    
    a) Padawan
    b) Jedi Master
    c) Force Ghost""")
        
    while (selected_level := input("\nChoice? "))  not in levels:
        print(f"Please select one of the options {', '.join(levels)}")
        
    return selected_level


def prepare_questions(questions, num_questions):
    num_questions = min(num_questions, len(questions))
    return random.sample(list(questions.items()), k=num_questions)

def ask_question(question, alternatives):
    correct_answer = alternatives[0]
    ordered_alternatives =random.sample(alternatives, k=len(alternatives))
    
    answer = get_answer(question, ordered_alternatives)
    if answer == correct_answer:
        print("\n ‚úÖ Correct! üéâ \n")
        return 1
    else:
        print(f"\n ‚ùå The answer is {correct_answer!r}, not {answer!r} \n")
        return 0

def get_answer(question,alternatives):
    print(f"{question}?")
    labeled_alternatives = dict(zip(ascii_lowercase, alternatives))
    for label, alternative in labeled_alternatives.items():
        print(f"    {label}) {alternative}")
        
    while (answer_label := input("\nChoice? ")) not in labeled_alternatives:
        print(f"Please answer one of {', '.join(labeled_alternatives)}")
        
    return labeled_alternatives[answer_label]
       

if __name__ == "__main__":
    run_quiz() 
        
```
